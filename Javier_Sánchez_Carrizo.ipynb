### Problema 1.- Funciones en Python (4 puntos)

##### Haz una función que imprima en pantalla un cuadrado que tenga un símbolo en las esquinas, otro distinto en las líneas exteriores (menos las esquinas) y otro distinto en la zona del centro.
##### Deberás permitir que el usuario introduzca con el teclado los símbolos que quiera y que indique el tamaño que debe tener el cuadrado.
##### NO PUEDES UTILIZAR NINGÚN MÓDULO
###### Nota: En la pantalla, el cuadrado se verá en forma de rectángulo al ser distinta la separación vertical y horizontal entre caracteres.
###### Ejemplo:
<img src='EjemploCuadrado.png' width=300px align=left>

```python
#Creamos dos variables que significaran los simbolos a introducir y el tamaño del cuadrado
simbolos=input("Introduce tres simbolos separados por comas")
tamaño=int(input("Introduce el tamaño"))
#Creamos la función a la que le pasamos los simbolos y el tamaño
def cuadrado(simbolos,tamaño):
    #Separamos los simbolos en un array llamado simbo mediante el método split pasandole como parámetro la ,
    simbo=simbolos.split(",")
    #Creamos una variable llamada lon que será la longitud del cuadrado restandole 2 (la primera y la última fila)
    lon=int(tamaño)-2
    ax=0
    #imprimimos la primera línea extrañendo del array el primer simbolo y el segundo las veces que mida el cuadrado -2
    print(f"{simbo[0]}{simbo[1]*(int(tamaño)-2)}{simbo[0]}")
    #hacemos un while para evaluar si nuestra variable aux(contador) es menor que la variable lon
    while(ax<lon):
        #Imprimimos el simbolo 2 y el 3 el tamaño del array -2
        print(f"{simbo[1]}{simbo[2]*(int(tamaño)-2)}{simbo[1]}")
        ax+=1
    #Por último imprimimos la última fila que será idéntica a la primera
    print(f"{simbo[0]}{simbo[1]*(int(tamaño)-2)}{simbo[0]}")
#llamamos a la función
cuadrado(simbolos,tamaño)
```

### Problema 2.-  Módulos - Acceso y Análisis de Datos en Python (2 puntos)

##### Utilizando el módulo Pandas:
##### - Carga el fichero .excel adjunto a un dataframe
##### - Guarda el nombre (celda A1 del excel) en una variable e imprímelo (color naranja de la imagen)
##### - Pon los nombres correspondientes a las columnas (Temperatura...) (color verde de la imagen)

##### - Extrae los datos de las columnas a series de pandas e imprímelos

<img src='ExplExcel.png' width=500px align=left>

```python
#Importamos el archivo pandas para poder rabajar con los dataframes, también importamos el archivo openpyxl para poder exportar un dataframe a un fichero excel
import pandas as pd
import openpyxl
#Leemos el fichero excel y lo almacenamos en un dataframe
df=pd.read_excel("Alto_Leon_enero_marzo_2021.xls")
print(df)
#Guardamos el nombre de la celda A1 en una variable y lo imprimimos
title=df.columns[0]
print(title)
#Cambiamos el nombre a las columnas por los correspondientes
df.rename(columns={"Puerto Alto del León":"Fecha y hora oficial","Unnamed: 1":"Temperatura Máxima(ºC)","Unnamed: 2":"Temperatura Mínima(ºC)","Unnamed: 3":"Temperatura Media(ºC)","Unnamed: 4":"Racha(Km/h)","Unnamed: 5":"Velocidad Máxima"},inplace=True)
print(df.columns)
print(df)
#Dropeamos del dataframe las filas 0,1 y 2, y lo almacenamos en un nuevo dataframe
nuevo=df.drop([0,1,2])
print(nuevo)
#Extraemos la columna Fecha y hora oficial y lo almacenamos en una Serie mediante el método Series()
serie=pd.Series(nuevo["Fecha y hora oficial"])
print(serie)
#Asignamos a los indices del dataframe los de la serie (es decir las fechas)
nuevo.index=serie
#Dropeamos la columna correspondiente a la de Fecha y hora oficial
n1=nuevo.drop(df.columns[0], axis='columns')
print(n1)

#Creamos series y le introducimos los valores de cada columna
s1=pd.Series(n1["Temperatura Máxima(ºC)"])
s2=pd.Series(n1["Temperatura Mínima(ºC)"])
s3=pd.Series(n1["Temperatura Media(ºC)"])
s4=pd.Series(n1["Racha(Km/h)"])
s5=pd.Series(n1["Velocidad Máxima"])

#Imprimimos las series
print(s1)
print("-------------")
print(s2)
print("-------------")
print(s3)
print("-------------")
print(s4)
print("-------------")
print(s5)
print("-------------")
#Extraemos las 10 primeras filas
print(nuevo.iloc[10])
#Exportamos el dataframe a un archivo excel llamado DatosExportados.xlsx a la hoja Datos.
n1.to_excel("DatosExportados.xlsx","Datos")
```

### Problema 4.-  Módulos - Acceso, Modificación y Consulta de Bases de Datos en Python (2 puntos)

##### Crea una base de datos con las siguientes tablas:
##### - cliente, con los campos idcliente, nombre, apellido1 y apellido2
##### - direccion, con los campos idcliente, calle, numero y municipio
##### - producto, con los campos idproducto, nombre y precio
##### - pedido, con los campos idpedido, idcliente, idproducto, fechapedido y fechaentrega
##### Introduce al menos tres valores para cada tabla
##### Genera al menos tres consultas distintas
###### Notas:
###### - Si pones las consultas en celdas separadas se imprimirá el resultado
###### - No te dejes la base de datos abierta, penaliza la nota

```python
#importamos la base de datos sqlite y creamos la base de datos mediante el método connect
import sqlite3 as sql
con=sql.connect("hito")
```

```python
#Creamos un cursor que ejecutará todas las sentencias que le demos
cur=con.cursor()
#Eliminamos las tablas de la bd (funcionará si está creada) ----> Si quieres borralas desomenta el bloque
# cur.execute("Drop table cliente")
# cur.execute("Drop table direccion")
# cur.execute("Drop table producto")
# cur.execute("Drop table pedido")
#Creamos las tablas cliente, direccion, producto y pedido
cur.execute("CREATE TABLE cliente(idcliente, nombre, apellido1,apellido2)")
cur.execute("CREATE TABLE direccion(idcliente, calle, numero,municipio)")
cur.execute("CREATE TABLE producto(idproducto, nombre, precio)")
cur.execute("CREATE TABLE pedido(idpedido, idcliente, idproducto,fechapedido,fechaentrega)")

#Introducimos los datos en cada tabla creando una lista con dichos datos y pasandola al executeany de las sentencias insert into
data = [
(1 , "Andres", "Sanchez","Huerta"),
(2 ,  "Andres", "Sanchez","Picaflor"),
(3 ,  "Andres", "Sanchez","Montero"),
]
cur.executemany("INSERT INTO cliente VALUES(?, ?, ?,?)" , data)


data2 = [
(1 , "Las margaritas", 8,"San fernano"),
(2 ,  "Pleistoceno", 56,"Sevilla la nueva"),
(3 ,  "Madrid", 3,"Getafe"),
]
cur.executemany("INSERT INTO direccion VALUES(?, ?, ?,?)" , data2)


data3 = [
(1 , "Tobogan", 38),
(2 ,  "Pistola de agua", 56),
(3 ,  "Cubo de la basura", 3),
]
cur.executemany("INSERT INTO producto VALUES(?, ?, ?)" , data3)


data4 = [
(1 , 1, 1,"07/04/2018","06/09/2020"),
(2 ,  2, 3,"12/05/2021","06/09/2022"),
(3 ,  3, 2,"12/04/2021","06/09/2022"),
]
cur.executemany("INSERT INTO pedido VALUES(?, ?, ?,?,?)" , data4)


#Guardamos los cambios realizados
con.commit()
```

```python
#Creamos 3 sentencias:
#Con la primera vemos todos los productos
for i in cur.execute("SELECT * from producto"):
    print(i)
#Con la segunda seleccionamos todos los clientes que se llamen Andres
for i in cur.execute("SELECT * from cliente where nombre='Andres'"):
    print(i)
#Con la tercera seleccionamos todas las calles cuyo numero sea 8
for i in cur.execute("SELECT * from direccion where numero=8"):
    print(i)
con.close()
```
